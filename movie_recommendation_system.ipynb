```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('/content/dataset.csv')
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies.info()
```

```python
movies['tags']=movies['genre']+movies['overview']
```

```python
new_df=movies[['id','genre','title','genre','overview','tags']]
```

```python
new_df =new_df.drop(columns=['genre','overview'])
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vec =cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sin=cosine_similarity(vec)
```

```python
sin
```

```python
new_df[new_df['title'] == 'The Godfather']
```

```python
dist =sorted(list(enumerate(sin[0])),reverse=True,key=lambda vec:vec[1])
```

```python
dist
```

```python
for i in dist[0:5]:
  print(new_df.iloc[i[0]].title)
```

```python
def recommend(movies):
  index=new_df[new_df['title']==movies].index[0]
  distance =sorted(list(enumerate(sin[index])),reverse=True,key=lambda vec:vec[1])
  for i in distance[0:5]:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```

```python
import pickle
with open('movie_recommender.pkl', 'wb') as f:
    pickle.dump(recommend, f)
```
